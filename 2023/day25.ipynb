```python
import networkx as nx

adj = {k:vs for k, vs in map(lambda pairs: (pairs[0].strip(), pairs[1].strip().split(" ")), map(lambda line: line.replace("\n", "").split(":"), open("problems/d25.txt", "r").readlines())) }


G = nx.Graph(adj)
G2 = G.copy()

for edge in nx.minimum_edge_cut(G):
    G2.remove_edge(edge[0], edge[1])

r = 1
for c in [len(c) for c in list(nx.connected_components(G2))]:
    r *= c

r
```

```python
import networkx as nx

adj = {k:vs for k, vs in map(lambda pairs: (pairs[0].strip(), pairs[1].strip().split(" ")), map(lambda line: line.replace("\n", "").split(":"), open("problems/d25.txt", "r").readlines())) }


G = nx.Graph(adj)
G2 = G.copy()

for edge in nx.minimum_edge_cut(G):
    G2.remove_edge(edge[0], edge[1])

r = 1
for c in [len(c) for c in list(nx.connected_components(G2))]:
    r *= c

r
```
